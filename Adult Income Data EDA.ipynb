```python
# Q1. Is there any effect of person's age on working hours?

# bivariate analysis - age and hours.per.week are both numerical columns of data.

sns.scatterplot(df['age'] , df['hours.per.week'])

plt.show()


# age and working hours of a person are not specifically related according to the sctterplot below. 
# However, at age of 90 , there is an extreme observation of 100 hours.per.week
```

```python
# Q2. Analyze trend of capital gain amount for different educational qualifications.

df[['capital.gain','education.num']].groupby('education.num').mean().plot()
plt.show()

# The highest average capital gain amount is for individuals with education number 15.
```

```python
# Q3. Check effect of working hours on capital gain.

# bivariate analysis - numerical, numerical columns

sns.lmplot(data = df, x='hours.per.week', y = 'capital.gain')
plt.show()

# LMplot i.e., Linear Model plot shows extreme points for capital gain of 1,00,000.
```

```python
# Q4. Find trend of average working hours corresponding to age of an individual.

df[['age', 'hours.per.week']].groupby('age').mean().plot()
plt.show()

# The lineplot shows that average working hours are increasing with age. After 60 years of age, working hours are decreasing. 
# It also shows that some older individuals with 80-90 years of age are working more than 30 hours per week.
```

```python
# Q5. Check distribution of average capital loss corresponding to age.

df[['age', 'capital.loss']].groupby('age').mean().plot()
plt.show()

# The lineplot shows that for individuals with age 80-85 years, there is sudden hike in capital loss.
```

```python
# Q6. Is there a negative relation between capital gain and age.

df[['age','capital.gain']].groupby('age').mean().plot()
plt.show()

# From the lineplot and scatterplot, we can be sure that there is no negative relation between age and capital.gain columns.
```

```python
# Q7. Compare average age with levels of income.

sns.barplot(df['age'] , df['income'])
plt.show()

# The barplot shows that income is greater than 50K for individuals above 40 years as people with higher age will have a higher experience and more income. 
```

```python
# Q8. check distribution of working hours with gender of individual

sns.violinplot(df['sex'] , df['hours.per.week'])
plt.show()

# For large proportion of females, working hours are between 20 to 50 hours per week. Males have longer working hours 40 to 60 hours per week.
```

```python
# Q9. compare income of individuals based on their education.

sns.boxplot(df['education.num'] , df['income'])
plt.show()

# The salary is greater than 50K for individuals with higher education number(10-13). Outliers are present in both categories with income <= 50K having more outliers.
```

```python
# Q11. Compare distibution of working hours for distinct classes of income.

sns.boxplot(df['income'] , df['hours.per.week'])
plt.show()

# There is difference between working hours based on income with outliers in both categories of income. Individuals with income greater than 50K are usually working more hours than <50K group.
```

```python
# Q12. Identify positively skewed distributions of age based on gender and race of individual.

# multivariate analysis - age, gender, race

sns.violinplot(data = df, x= 'age' , y= 'race' , hue ='sex' , split ='sex')
plt.show()

# The violinplot shows Positive Distribution for (White, Female), (Asian-Pac-Islander, Female), (Other, Female), (Other, Male), (Amer-Indian-Eskimo, Male).
```

```python
# Q13. Compare the working hours for each working class.

plt.figure(figsize = (18,7))
sns.boxplot(data = df, x='workclass' , y='hours.per.week')
plt.show()

# We can infer that high proportion of the federal-gov employees have nearly 40 working hours per week. But, there are some federal-gov employees with working hours other than 40. Also, the working hours for individuals varies greatly depending upon their professions.
```

```python
# Q14. compare the weightage based on demographic characteristics by gender. 

sns.boxplot(data = df , x='sex' , y ='fnlwgt')
plt.show()

# gender does not impact weightage based on demographic characteristics variable.
```

```python
# Q15. Study variation in income by gender.

pd.crosstab(df['sex'] , df['income'] ).plot(kind = 'bar' , stacked = True)
plt.show()

# Approximately 20% of males have high income > 50K compared to females.
```

```python
# Q16. compare income by race.


pd.crosstab(df['race'] , df['income'] ).plot(kind = 'bar' , stacked = True, figsize = (18,7) )
plt.show()

# White people have very high income compared to other categories in race column.
```

```python
# Q17. Is there an all male profession according to data.

pd.crosstab(df['occupation'] , df['sex']).plot(kind = 'bar' , stacked = True, figsize = (20,8) )
plt.show()

# Total 9 rows have occupation as 'Armed-Forces' and none of these 9 rows have 'sex' value as ' Female' .
# Hence, 'Armed-Forces' is shown as all Male profession.
```

```python
# Q18. How does income vary with working class ?

pd.crosstab( df['workclass'] , df['income'] ).plot( kind= 'bar' , figsize = (18,7) , stacked = True )
plt.show()

# Private working class people have higher incomer and most frequent working class within individuals represented in the data.
```

```python
# Q19. How is income related to marital status?
    
pd.crosstab(df['income'] , df['marital.status'] ).plot( kind = 'bar' , figsize = (18,7) )
plt.ylabel('income')
plt.xlabel('marital status')
plt.title('Income relation with Marital status')
plt.show()

# Individuals with 'marital.status' as 'Never-married ' have income <= 50K mostly possibly due to less age and experience.
# 'Married-civ-spouse' have highest incomes in both groups of <=50K and >50K with mostly in <=50K category.
```

```python
# Q20. Compare income for different occupations.

pd.crosstab(df['income'] , df['occupation'] ).plot(kind = 'bar' , stacked = True  , figsize = (18,7))
plt.show()

# Individuals with income >50K mostly have occupations as 'Prof-specialty' and 'Transport-moving' .
```

```python
df['occupation'].value_counts().plot(kind = 'pie' , figsize= (18,7) , autopct ='%i%%' )

plt.show()
```

```python
count = pd.pivot_table(df,index=["occupation"], columns=["income"], values='age', aggfunc='mean')

fig = count.plot.bar(figsize=(15,8))

plt.title('Comparison of Income for Different Occupations', fontsize = 15)
fig.set_xlabel('Occupation', fontsize = 15)
fig.set_ylabel('Age', fontsize = 15)

fig.legend (loc ='upper right')
plt.show()
```
