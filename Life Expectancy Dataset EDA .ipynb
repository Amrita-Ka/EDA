```python
#Q8. Define a function to find extreme values in alcohol consumption.

sns.boxplot(df['Alcohol'])
plt.show()

# Boxplot shows presence of outliers in Alcohol column.
```

```python
# function to find exact values.

#threshold = 3 is set as per z-score metric

outliers = []
def detect_outlier(data):
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    
    for y in data:
        z_score = (y - mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers

outlier_pt = detect_outlier(df.Alcohol)

print('Extreme values for alcohol:\n' + str(outlier_pt))

# The 3 values in output list are the extreme values in Alcohol consumption column.
```

```python
#Q9. Create a column 'Life_expectancy_level' such that it will contain 3 levels (High, Medium, Low) based on life expectancy.

df['Life expectancy'].describe()
```

```python
age = df['Life expectancy']

def feature(age):
    if age < df['Life expectancy'].quantile(0.25):
        return 'Low'
    if (age >= df['Life expectancy'].quantile(0.25))&((age<df['Life expectancy'].quantile(0.75))):
        return 'Medium'
    if (age > df['Life expectancy'].quantile(0.75)):
        return 'High'
    

df['Life_expectancy_level'] = df['Life expectancy'].apply(feature)
    
```

```python
df.head()

# We created a new column as 'Life_expectancy_level' that categorises 'Life expectancy' into 3 levels Low, Medium, High.
```
