```python
# Q13. create dummy variable for each category in region.

# One-hot encoding of the 3 category columns - gender, region, smoker.

# df1 = pd.get_dummies(data = df , columns= ['region' , 'gender', 'smoker'] , drop_first =False)

df = pd.get_dummies(data = df , columns= ['region'] , drop_first =False)

df.head()
```

```python
# Q14. create variable smoker_label containing unique label for each category in variable 'smoker'

from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()
df['smoker_label'] = lab.fit_transform(df['smoker'])
```

```python
# Q15. Create variable which will contain frequency of an individual's gender.

# frequency encoder for gender

encoding = df.groupby('gender').size()
encoding_freq = encoding/len(df)
df['freq_encoded_gender'] = df.gender.map(encoding_freq)
df.head()
```

```python
# Q16. create a variable such that 1 indicates smoker and 0 non-smoker.

df['smoker'] = df['smoker'].replace({'smoker':1 , 'non-smoker':0})

df.head()
```

```python
#Q19. Check skewness for 'medical_cost'. If data is skewed, use appropriate technique to reduce skewness in data.

plt.hist(df['medical_cost'] , bins=100)

plt.show()

# right skewed.
```

```python
df.medical_cost.skew()

# positive skewness, right - skewed data in medical_cost column.

# Square-root and log transformation both are effective to transform data if skewed.
```

```python
# log transformation of medical_cost column.

df['log_medical_cost'] = np.log(df['medical_cost'])
plt.hist(df['log_medical_cost'] , bins =100)
plt.show()
```

```python
df['log_medical_cost'].skew()

# the skewness of log transformed medical_column data is near zero so the method has effectively reduced skewness in data.
```

```python
# Q20. Transform transformed_medical_cost to original scale and compare medical_cost and transformed data.

# exponential method implementation

df['exp_medical_cost'] = np.exp(df['log_medical_cost'])

print(df[['medical_cost', 'exp_medical_cost']].describe())
```

```python
plt.figure(figsize=(15,8))

plt.subplot(1,2,1)
sns.distplot(df['medical_cost'])


plt.subplot(1,2,2)
sns.distplot(df['exp_medical_cost'])

plt.show()
```

```python
#Comparison shows that the exponential transformation of 'log_medical_cost' transforms data to original scale of 'medical_cost'.
```
